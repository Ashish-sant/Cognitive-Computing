```python
X= ord('A') +ord('S')
sales = np.array([X,X+50,X+100,X+150,X+200])
print(sales)
```

```python
tax=((X%5)+5)/100
print(tax)
```

```python
ta = np.array([0.8,0.8,0.8,0.8,0.8])
taxed = sales*ta;
print(taxed)
```

```python
discount = np.where(sales<X+100, sales*0.05,0)
discount
```

```python
discount = np.where(sales<X+100, sales*0.1,0)
discount
```

```python
week = np.vstack([sales,sales*0.02,sales*0.04])
print(week)
```

```python
x = np.linspace(-10,10,100)
y1=x*x
y2=np.sin(x)
y3=np.exp(x)
y4=np.log(np.abs(x)+1)

import matplotlib.pyplot as plt
plt.plot(x,y1)
plt.plot(x,y2)
# plt.plot(x,y3)
plt.plot(x,y4)
plt.title('Sample graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.legend(['y1','y2','y3','y4'])
plt.show()
```
