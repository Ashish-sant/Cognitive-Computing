```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = float(input("Enter m value: "))
x = np.linspace(-10,10,100)
h = m*x**2
g = m*np.sin(x)
plt.plot(x,h,'b--', label="h(x)")
plt.plot(x,g,'r--' , label="g(x)")
plt.legend()
plt.grid()
plt.title("Math Functions")
plt.show()
```

```python
import pandas as pd
data = {
    'Subject': ['Math', 'Physics', 'Chemistry', 'Biology', 'Computer Science'],
    'Score': [85, 90, 78, 88, 95]
}
df = pd.DataFrame(data)
print(df)
```

```python
import seaborn as sns
colors = ['red', 'blue', 'green', 'orange', 'purple']
bars = sns.barplot(x='Subject', y='Score', data=df, palette=colors)
plt.grid()
plt.title("Subject Scores")
for i in bars.containers:
  bars.bar_label(i)
plt.show()
```

```python
Roll = int(input("Enter rollnumber"))
np.random.seed(Roll)
x = np.random.randn(50)
print(x)
```

```python
fig, axes = plt.subplots(2,2)
axes[0, 0].plot(np.cumsum(x), linestyle='-', color='b')
axes[0, 0].set_title("Cumulative Sum Line Plot")
axes[0, 0].set_xlabel("Index")
axes[0, 0].set_ylabel("Cumulative Sum")
axes[0, 0].grid()

axes[0, 1].scatter(range(50), x, color='r')
axes[0, 1].set_title("Scatter Plot with Noise")
axes[0, 1].set_xlabel("Index")
axes[0, 1].set_ylabel("Random Value")
axes[0, 1].grid()
```

```python
url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)
```

```python
sns.lineplot(x =df['month_number'] , y=df['total_profit'],color='b')
plt.title("Total profits per month")
plt.xlabel("Month")
plt.ylabel("Total profit")
plt.grid()
plt.show()

for i in df.columns[1:-1]:
    sns.lineplot(x=df['month_number'], y=df[i], label=i)
plt.title("Product Sales Over Months")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.grid()
plt.show()

df.plot(kind='bar', figsize=(12, 6))
plt.title("Bar Chart of Sales Data")
plt.xlabel("Month")
plt.ylabel("Values")
plt.legend()
plt.grid()
plt.show()
```
