```python
#list of 100 random numbers
import random as r
l = [r.randint(100,900) for i in range(100)]
even = 0
odd=0
for i in l :
  if(i%2==0):
    even += 1
  else:
    odd +=1
print(even)
print(odd)
```

```python
#prime numbers
for i in l:
  for j in range(2,i):
    if(i%j==0):
      break
  else:
    print(i)
```
