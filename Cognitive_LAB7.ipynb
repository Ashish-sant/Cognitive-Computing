```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
roll = int(input("Enter roll number: "))
np.random.seed(roll)
```

image.png

```python
sales_data = np.random.randint(1000, 5001, size=(12, 4))
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
columns = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
df = pd.DataFrame(sales_data, columns=columns, index=months)
```

```python
print("Sales Data:")
print(df)
```

image.png

```python
print("First 5 rows:")
print(df.head())
print("Summary Statistics:")
print(df.describe())
```

```python
total_sales_cat = df.sum()
print("\nTotal Sales category:")
print(total_sales_cat)

total_sales_mon = df.sum(axis=1)
print("Total Sales Month:")
print(total_sales_mon)
```

```python
sales_growth = df.diff().mean()
print("Average Sales Growth between Consecutive Months:")
print(sales_growth)
```

```python
df['Total Sales'] = df.sum(axis=1)

df['Growth Rate'] = df['Total Sales'].pct_change() * 100
print("DataFrame with Total Sales and Growth Rate:")
print(df)
```

```python
df_discounted = df.copy()
if roll % 2 == 0:
    df_discounted['Electronics'] = df_discounted['Electronics'] * 0.9
    print("Applied 10% discount on Electronics (Even Roll).")
else:
    df_discounted['Clothing'] = df_discounted['Clothing'] * 0.85
    print("Applied 15% discount on Clothing (Odd Roll).")
print("Data after Discount:")
print(df_discounted)
```

image.png

```python
for category in columns:
    plt.plot(df.index, df[category], marker='o', label=category)
plt.title("Monthly Sales Trends by Category")
plt.xlabel("Month")
plt.ylabel("Sales (Units)")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[columns])
plt.title("Sales Distribution for Each Category")
plt.xlabel("Product Category")
plt.ylabel("Sales (Units)")
plt.grid()
plt.show()
```

```python
import numpy as np

array = np.array([[1, -2, 3],
                  [-4, 5, -6]])
```

```python
abs_array = np.abs(array)
print("Element-wise absolute value:")
print(abs_array)
```

```python
flat = np.percentile(array.flatten(), [25, 50, 75])
columns = np.percentile(array, [25, 50, 75], axis=0)
rows = np.percentile(array, [25, 50, 75], axis=1)

print("Percentiles of flattened array (25th, 50th, 75th):")
print(flat)

print("Percentiles for each column (rows=[25,50,75]):")
print(columns)

print("Percentiles for each row (rows=[25,50,75]):")
print(rows)
```

```python
mean_flat = np.mean(array.flatten())
median_flat = np.median(array.flatten())
std_flat = np.std(array.flatten())

mean_columns = np.mean(array, axis=0)
median_columns = np.median(array, axis=0)
std_columns = np.std(array, axis=0)

mean_rows = np.mean(array, axis=1)
median_rows = np.median(array, axis=1)
std_rows = np.std(array, axis=1)

print("Flattened array: Mean =", mean_flat, "Median =", median_flat, "Std Dev =", std_flat)
print("Per column: Mean =", mean_columns, "Median =", median_columns, "Std Dev =", std_columns)
print("Per row: Mean =", mean_rows, "Median =", median_rows, "Std Dev =", std_rows)
```

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_a = np.floor(a)
ceil_a = np.ceil(a)
trunc_a = np.trunc(a)
round_a = np.round(a)

print("Original array:")
print(a)

print("Floor values:")
print(floor_a)

print("Ceiling values:")
print(ceil_a)

print("Truncated values:")
print(trunc_a)

print("Rounded values:")
print(round_a)
```
